```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(2, 1, figsize=(13, 13), gridspec_kw={'height_ratios': [7, 7]})


sns.boxplot(data=data1["Newspaper"], ax=axes[0], color='skyblue', width=0.5)
axes[0].set_title("Box plot")
axes[0].set_xlabel("Box plot")
```

```python
sns.histplot(data1["Newspaper"], kde=True, ax=axes[1], color='purple', bins=30)
axes[1].set_title("Histogram with KDE")
axes[1].set_xlabel("Newspaper")
axes[1].set_ylabel("Frequency")


plt.tight_layout()

plt.show()
```

Observation

    A high corelation is observed between daily and sunday

```python
x = data1["daily"].values
y = data1["sunday"].values
plt.scatter(x, y, color = "m", marker = "o", s = 30)
b0 = 13.84
b1 =1.33
# predicted response vector
y_hat = b0 + b1*x
 
# plotting the regression line
plt.plot(x, y_hat, color = "g")
  
# putting labels
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
